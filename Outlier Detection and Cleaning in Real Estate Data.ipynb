```python
plt.hist(df.price_per_sqft , bins=20 ,rwidth=0.8)
plt.xlabel('price per sqft')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df.price_per_sqft , bins=20 ,rwidth=0.8)
plt.xlabel('price per sqft')
plt.ylabel('Count')
plt.yscale('log')
plt.show()
```

```python
max_thershold = df['price_per_sqft'].quantile(0.999)
max_thershold
```

```python
min_thershold = df['price_per_sqft'].quantile(0.001)
min_thershold
```

SHOWING OUTLINER 

```python
df[(df.price_per_sqft>max_thershold) | (df.price_per_sqft<min_thershold)]
```

```python
data = df[(df.price_per_sqft<max_thershold) & (df.price_per_sqft>min_thershold)]
data.shape
```

```python
df.shape[0] - data.shape[0]
```

```python
df2 = data
```

```python
df2.sample(10)
```

```python
upper_limt = df2.price_per_sqft.mean() + 4*df2.price_per_sqft.std()
upper_limt
```

```python
lower_limit =  df2.price_per_sqft.mean() - 4*df2.price_per_sqft.std()
lower_limit
```

```python
df2[(df2.price_per_sqft>upper_limt) | (df2.price_per_sqft<lower_limit)]
```

```python
df3 = df2[(df2.price_per_sqft<upper_limt) & (df2.price_per_sqft>lower_limit)]
df3.tail(40)
```

```python
df2.shape[0] - df3.shape[0]
```

```python
from scipy.stats import norm
import numpy as np
```

```python
plt.hist(df3.price_per_sqft,bins=20,rwidth=0.8,density=True)
plt.xlabel('price_per_sqft')
plt.ylabel('count')
plt.show()

```

```python
plt.hist(df3.price_per_sqft,bins=20,rwidth=0.8,density=True)
plt.xlabel('price_per_sqft')
plt.ylabel('count')
plt.show()

rng = np.arange(-5000,df3.price_per_sqft.max(),100)
plt.plot(rng,norm.pdf(rng,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))
```

```python
final_data = df3[~df3['location'].isin(['other', 'unknown'])]
```

```python
final = final_data[final_data['location'] != 'other'].reset_index(drop=True) #reorded
```

```python
final.to_csv("cleaned_data.csv", index=False)
```
